### Ensemble technique

#### Random Forest

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import KFold
```

```python
k_fold=KFold(n_splits=5,shuffle=False,random_state=100)
```

```python
f1=['perimeter_mean','compactness_mean','compactness_mean','radius_worst','perimeter_worst','points_worst']
```

```python
t1=['diagnosis']
```

```python
knn_k_vals=[1,3,5]
```

```python
avg_train_acc=[]
avg_val_accuracy=[]
```

```python
X=bc_df[f1]
```

```python
y=bc_df[t1]
```

```python
for k in knn_k_vals:
    train_acc=[]
    val_acc=[]
    for i,(c_train,c_test) in enumerate(k_fold.split(X,y)):
        model=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
        model.fit(X[f1],y[t1])
        train_acc.append(model.score(X.iloc[c_train],y.iloc[c_train]))
        val_acc.append(model.score(X.iloc[c_test],y.iloc[c_test]))
    avg_train_acc.append(np.mean(train_acc))
    avg_val_acc.append(np.mean(val_acc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=10,min_samples_split=20,min_impurity_decrease=0.05)
model.fit(c_train[f1],c_train[t1])
```

```python
train_acc=model.score(c_train[f1],c_train[t1])
test_acc=model.score(c_test[f1],c_test[t1])
```

```python
print(train_acc,test_acc)
```

```python
#Feature Importance
feature_imps=pd.DataFrame(np.array([f1,model.feature_importances_]).T,columns=["feature","importance"])
feature_imps.sort_values(by="importance",ascending=False)
```
